```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
dataset['milage']=dataset['highwaympg']+dataset['citympg']
```

```python
def check_vif(n):
    vif=pd.DataFrame()
    vif['variables']=n.columns
    vif['VIF']=[variance_inflation_factor(n.values,i) for i in range(n.shape[1])]
    return vif
    
```

```python
check_vif(dataset[[i for i in dataset.describe().columns if i not in ['peakrpm','curbweight','stroke','carwidth','carheight','price','car_ID','citympg','highwaympg','wheelbase', 'carlength','boreratio', 'horsepower']]])
```

```python
df2['cyliender_No']=dataset['cylindernumber'].replace({'four':4,'six':6,'eight':8,'five':5,'two':2,'three':3,'twelve':12})
```

```python
df2['door_No']=dataset['doornumber'].replace({'four':4,'two':2})
```

```python
df2['price']=dataset['price']
```

```python
cor=df2.corr()
sns.heatmap(abs(cor),annot=True,cmap='coolwarm')
```

```python
df2.drop(['door_No'],axis=1,inplace=True)
```

```python
df2['CarName']=dataset['CarName'].str.split(" ",expand=True)[0]
```

```python
df2['CarName'].unique()
```

```python
df2['CarName']=df2['CarName'].replace({'vokswagen':'volkswagen','vw':'volkswagen','toyouta':'toyota','nissan':'Nissan','porsche':'porcshce'})
```

```python
carname=df2.groupby(['CarName'])['price'].mean().sort_values(ascending=False)
carname.plot.bar()
```

```python
df2['category']=df2['price'].apply(lambda x: 0 if x<15000 else 1 if 15000<x<25000 else 2)
```

```python
#df2=pd.get_dummies(dataset,columns=['fueltype','fuelsystem','enginetype','drivewheel','carbody','aspiration'])
```

```python
df2['fueltype']=dataset['fueltype']
df2['fuelsystem']=dataset['fuelsystem']
df2['enginetype']=dataset['enginetype']
df2['drivewheel']=dataset['drivewheel']
df2['carbody']=dataset['carbody']
df2['aspiration']=dataset['aspiration']
```

```python
df2=pd.get_dummies(df2,columns=['category','fueltype','fuelsystem','enginetype','drivewheel','carbody','aspiration'])
```

```python
df2
```

```python
df2.drop(['CarName'],axis=1,inplace=True)
```

```python
for i in df2.columns[0:5]:
    plt.boxplot(df2[i])
    plt.title(i)
    plt.show()
```

```python
df2.drop(df2[(df2['milage']>90)].index,axis=0,inplace=True)
```

```python
df2.head()
```

```python
df2.shape
```

```python
independent_variable=['symboling', 'compressionratio', 'enginesize', 'milage', 'cyliender_No',
       'category_0', 'category_1', 'category_2', 'fueltype_diesel',
       'fueltype_gas', 'fuelsystem_1bbl', 'fuelsystem_2bbl', 'fuelsystem_4bbl',
       'fuelsystem_idi', 'fuelsystem_mfi', 'fuelsystem_mpfi',
       'fuelsystem_spdi', 'fuelsystem_spfi', 'enginetype_dohc',
       'enginetype_dohcv', 'enginetype_l', 'enginetype_ohc', 'enginetype_ohcf',
       'enginetype_ohcv', 'enginetype_rotor', 'drivewheel_4wd',
       'drivewheel_fwd', 'drivewheel_rwd', 'carbody_convertible',
       'carbody_hardtop', 'carbody_hatchback', 'carbody_sedan',
       'carbody_wagon', 'aspiration_std', 'aspiration_turbo']
```

```python
Y=df2['price']
```

```python
X=df2[independent_variable]
```

```python
from scipy.stats import zscore
X=X.apply(zscore)
```

```python
Y=np.log10(Y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=50)
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
model.score(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```

```python
X_train_predicted=model.predict(X_train)
r2_score(Y_train,X_train_predicted)
```

```python
mean_squared_error(Y_train,X_train_predicted)
```

```python
np.sqrt(mean_squared_error(Y_train,X_train_predicted))
```

```python
r2=r2_score(Y_train,X_train_predicted)
float(r2)
```

```python

Ar=1-((1.0-float(r2))*(202-1)/(202-35-1))
```

```python
print('***********summary of the model*****************')
print(' MSE :',mean_squared_error(Y_train,X_train_predicted))
print('RMSE :',np.sqrt(mean_squared_error(Y_train,X_train_predicted)))
print(' MAE :',mean_absolute_error(Y_train,X_train_predicted))
print('MAPE :',mean_absolute_percentage_error(Y_train,X_train_predicted))
print(' R^2 :',r2_score(Y_train,X_train_predicted))
print('AR^2 :',Ar)
```
